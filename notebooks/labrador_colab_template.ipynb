```python
import re
from pathlib import Path

path = Path("/content/PyHealth_for_labrador/pyhealth/models/labrador.py")
text = path.read_text()

pattern = r'''result = \{"logit": logit, "y_prob": y_prob\}\s+if labels is not None:\s+result\["loss"\] = self\.get_loss_function\(\)\(logit, labels\)\s+result\["y_true"\] = labels\s+return result'''

replacement = '''result = {"logit": logit, "y_prob": y_prob}

        if labels is not None:
            if self.mode == "binary":
                labels = labels.float().view_as(logit)
            elif self.mode == "multilabel":
                labels = labels.float().view_as(logit)

            result["loss"] = self.get_loss_function()(logit, labels)
            result["y_true"] = labels

        return result'''

new_text, count = re.subn(pattern, replacement, text, flags=re.MULTILINE)
print("replacements:", count)

if count != 1:
    raise ValueError("替换失败，请把 159-170 行再次贴给我。")

path.write_text(new_text)
print("patched labrador.py")
```

```python
from pathlib import Path

path = Path("/content/PyHealth_for_labrador/tests/test_labrador.py")
text = path.read_text()

text = text.replace('assert out["logit"].shape == (3, 2)', 'assert out["logit"].shape == (3, 1)')
text = text.replace('assert out["y_prob"].shape == (3, 2)', 'assert out["y_prob"].shape == (3, 1)')
text = text.replace('assert out["logit"].shape == (2, 2)', 'assert out["logit"].shape == (2, 1)')
text = text.replace('dataset.get_output_size = mocker.MagicMock(return_value=2)', 'dataset.get_output_size = mocker.MagicMock(return_value=1)')

if 'assert out["y_true"].shape == (3, 1)' not in text:
    text = text.replace(
        'assert out["y_prob"].shape == (3, 1)',
        'assert out["y_prob"].shape == (3, 1)\n        assert out["y_true"].shape == (3, 1)'
    )

path.write_text(text)
print("patched tests/test_labrador.py")
```
